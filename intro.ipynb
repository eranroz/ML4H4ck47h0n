# Data

First we have to get ***data***!

Here we will use 2 datasets:

* Results from Knesset elections which are accessible from <http://www.bechirot20.gov.il/>
* Ishuvim details from Central Bureau of Statistics (למ"ס) <http://www.cbs.gov.il/ishuvim/ishuvim_main.htm>

```python
import requests
req_elections_data = requests.get('http://www.votes20-media.gov.il/files/expb.csv')
with open('elections20.csv', 'w') as elections_fd:
    # for god sake it's 2015 who use windows-1255?! we will use unicode...
    unicode_elections_data = req_elections_data.content.decode('windows-1255')
    elections_fd.write(unicode_elections_data)

lamas_data = requests.get('http://www.cbs.gov.il/ishuvim/ishuv2014/bycode.xls')
with open('lamas.xls','wb') as lamas_fd:
    lamas_fd.write(lamas_data.content)
rashuyot_req = requests.get('http://www.cbs.gov.il/publications15/local_authorities13_1609/excel/p_libud.xls')
with open('rashuyot.xls','wb') as rashuyot_fd:
    rashuyot_fd.write(rashuyot_req.content)
```

Let's take a first look on the data we have. I'm going to use [pandas](http://pandas.pydata.org/), a great library for queries and data analysis

```python
import pandas as pd
from IPython.display import display
%matplotlib inline
import matplotlib.pyplot as plt
# some settings to have nice display (you can put it in config file somewhere outside)
plt.rcParams['font.family'] = ['Arial']
plt.rcParams['figure.figsize'] = (12.0, 12.0)
plt.style.use('ggplot')
pd.options.display.max_rows = 10
# end of nice settings

votes = pd.read_csv('elections20.csv')
votes.columns = [c.strip() for c in votes.columns]  # clean whitespaces
parties = votes.columns[7:]

votes = votes.loc[votes[parties].sum(1) > 0]  # remove rows with no votes
# display the data    
display(votes)

# summary the data using pandas + some basic plots
votes.sum()[7:].plot(kind='pie', labels=[p[::-1] for p in parties], fontsize=16)
```

as you can see pandas is awesome! Similar to SQL we can groupby and aggregate efficently and simple.
For example if we are interested in the number of votes per city and not per kalpi we can use group by.

We can also get a sorted list of parties, and include only real parties (שעברו את אחוז החסימה).

```python
votes_by_city = votes.groupby('שם ישוב').sum()
display(votes_by_city)

# to make the analysis easier (mainly for infographics reasons...)
parties_sorted = parties[votes.sum()[7:].argsort()[::-1]]
real_parties = parties_sorted[:10]
display(parties_sorted)
```

# Machine learning for data visualization and more

We got more than 10 thousands samples (קלפיות) and for each sample we have features - e.g properties that describes it - in our case: number of votes. So we have a lot of data, how can we visualize it? are there intersting patterns there?

We will use the popular machine learning package [sklearn/scikit-learn](http://scikit-learn.org)

## PCA - Dimensionality reduction

So we got at least 10 "dimensions" - where each dimensions represent a party. How can we visualize 10 dimension space? A common method to do this is [PCA](https://en.wikipedia.org/wiki/Principal_component_analysis)

```python
from sklearn.decomposition import PCA
votes[parties] = votes[parties].div(votes[parties].sum(1), axis=0)  # normalize rows
pca = PCA(n_components=2).fit(votes[real_parties].values)

max_part = votes[real_parties].idxmax(1)

plt.title('קלפיות - צבעים לפי המפלגה שקיבלה הכי הרבה בקלפי'[::-1])
plt.xticks([])
plt.yticks([])

cmap = plt.cm.get_cmap('Dark2', len(real_parties))
colors = [cmap(i) for i, _ in enumerate(real_parties)]

for i, p in enumerate(real_parties[pca.components_[0, :].argsort()[::-1]]):
    plt.scatter(*(pca.transform(votes.loc[max_part==p][real_parties].values).T), label=p[::-1], c=colors[i])

plt.legend(loc='upper left', ncol=3)
```

This is really cool - we have nice visualization where we can easily see "left" and what is "right" and similar parties are placed closed! 

# K-nearest neighbors

We can use a model called KNN - which assigns labels for points based on the K most closed points which we already know their labels. This is sometimes useful as visualization trick:


```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
pca_xy = pca.transform(votes[real_parties].values)
min_xy, max_xy = np.min(pca_xy, 0), np.max(pca_xy, 0)
c_pred=votes[real_parties].values.argmax(1)
c_pred = pca.components_[0, :].argsort()[::-1][c_pred]
knn = KNeighborsClassifier(1).fit(pca_xy, c_pred)
h = 0.01
xx, yy = np.meshgrid(np.arange(min_xy[0], max_xy[0], h),
                         np.arange(min_xy[1], max_xy[1], h))
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])

plt.title('קלפיות - צבעי רקע לפי הקלפי הדומה ביותר'[::-1])
plt.xticks([])
plt.yticks([])
plt.pcolormesh(xx, yy, Z.reshape(xx.shape), cmap=plt.cm.get_cmap('Pastel1', len(real_parties)))
plt.scatter(*pca_xy.T, marker='.', c='black', alpha=0.5)
```

PCA and KNN are very powerful - we can use visualization with interesting data to further study an patterns such as turnout (שיעור הצבעה):

```python
plt.title('שיעור הצבעה'[::-1])
plt.xticks([])
plt.yticks([])

plt.pcolormesh(xx, yy, Z.reshape(xx.shape), cmap=plt.cm.get_cmap('Pastel1', len(real_parties))) 
plt.scatter(*(pca.transform(votes[real_parties].values).T), c=votes['מצביעים']/votes['בזב'])
plt.colorbar()
```

# Supervised learning: regression and classification

Two common predictions problems in machine learning.

* Classification - Given some properties of a sample (features), assign it to a class (give it a label)
* Regression - Predict unkown variable

We give the algorithm training set - set of samples with the correct labels (or with the variable we measured) and train it on them to get a model, which can be used to predict future examples.

In our case we will use the votes in kalpis to predict details on the city - socio-economic rank:
